```python
x = df.drop(['price', 'CarName', 'fueltype', 'aspiration',	'doornumber', 'carbody',	'drivewheel', 'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem'], axis=1)
x
```

```python
y = df['price']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1012)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_train)
y_pred
```

```python
errorsq = metrics.r2_score(y_train, y_pred)
print("Error square: ", errorsq)
```

```python
plt.scatter(y_train,y_pred)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title("Actual Price vs Predicted Price")
plt.show()
```

```python
model = Lasso()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_train)
y_pred
```

```python
errorsqq = metrics.r2_score(y_train, y_pred)
print("Error square: ", errorsqq)
```

```python
plt.scatter(y_train, y_pred)
plt.xlabel('Actual price')
plt.ylabel('Predicted price')
plt.title("Actual vs Predicted")
plt.show()
```
