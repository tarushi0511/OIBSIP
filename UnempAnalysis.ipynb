```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import warnings
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month_name()
df['Year'] = df['Date'].dt.year
```

```python
warnings.filterwarnings("ignore")
df
```

```python
sns.countplot(x = df['Region'])
plt.xticks(rotation ='vertical')
plt.show()
```

```python
plt.figure(figsize = (3, 2))
sns.countplot(x = df['Area'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize = (3,2))
sns.heatmap(df[['Estimated Unemployment Rate (%)',
                'Estimated Employed',
                'Estimated Labour Participation Rate (%)']].corr())
plt.show()
```

```python
unemploment = df[["Area", "Region", "Estimated Unemployment Rate (%)"]]
fig = px.sunburst(unemploment, path = ["Area", "Region"],
                     values = "Estimated Unemployment Rate (%)",
                     width = 700, height=700,
                     title ="Unemployment Rate (Region)")
fig.show()
```

```python
fig = px.bar(df, x = 'Region', y = "Estimated Unemployment Rate (%)", color = "Region", title = "Avg Unemploment Rate")
fig.update_layout(xaxis = {'categoryorder':'total descending'})
fig.show()
```

```python
fig = px.bar(df, x = 'Month', y = 'Estimated Employed', color = 'Month', title = 'Est Employed')
fig.show()
```

```python
px.scatter(data_frame = df, x = 'Estimated Employed', y = 'Region')
```

```python
px.scatter(data_frame = df, x = 'Estimated Unemployment Rate (%)', y = 'Region')
```

```python
px.scatter(data_frame = df, x = 'Estimated Labour Participation Rate (%)', y = 'Region')
```
